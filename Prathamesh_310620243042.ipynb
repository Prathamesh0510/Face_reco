```python
len_of_images_cillian=len(cillian_gray)
```

```python
len_of_images_matt=len(matt_gray)
```

```python
len_of_images_virat=len(virat_gray)
```

```python
image_size_cillian = cillian_gray[9].shape
image_size_cillian
```

```python
image_size_matt = matt_gray[40].shape
image_size_matt
```

```python
image_size_virat = virat_gray[29].shape
image_size_virat
```

```python
flatten_size_cillian=image_size_cillian[0]*image_size_cillian[1]
flatten_size_cillian
```

```python
flatten_size_matt=image_size_matt[0]*image_size_matt[1]
flatten_size_matt
```

```python
flatten_size_virat=image_size_virat[0]*image_size_virat[1]
flatten_size_virat
```

```python
for i in range(len_of_images_cillian):
    cillian_gray[i] = np.ndarray.flatten(cillian_gray[i]).reshape(flatten_size_cillian,1)
```

```python
cillian_gray = np.dstack(cillian_gray)
cillian_gray = np.rollaxis(cillian_gray, axis=2, start=0)
cillian_gray.shape
```

```python
cillian_gray = cillian_gray.reshape(len_of_images_cillian, flatten_size_cillian)
cillian_gray.shape
```

```python
for i in range(len_of_images_matt):
    matt_gray[i] = np.ndarray.flatten(matt_gray[i]).reshape(flatten_size_matt,1)
```

```python
matt_gray = np.dstack(matt_gray)
matt_gray = np.rollaxis(matt_gray, axis=2, start=0)
matt_gray.shape
```

```python
matt_gray = matt_gray.reshape(len_of_images_matt, flatten_size_matt)
matt_gray.shape
```

```python
for i in range(len_of_images_virat):
    virat_gray[i] = np.ndarray.flatten(virat_gray[i]).reshape(flatten_size_virat,1)
```

```python
virat_gray = np.dstack(virat_gray)
virat_gray = np.rollaxis(virat_gray, axis=2, start=0)
virat_gray.shape
```

```python
virat_gray = virat_gray.reshape(len_of_images_virat, flatten_size_virat)
virat_gray.shape
```

```python
from sklearn.utils import shuffle
```

```python
images_indexed = shuffle(combine_2).reset_index()
```

```python
images_indexed
```

```python
drop_index = images_indexed.drop(['index'],axis =1)
```

```python
drop_index
```

```python
x = drop_index.values[:,:-1]
```

```python
y = drop_index.values[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn import svm
```

```python
clf = svm.SVC()
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn import metrics
```

```python
accuracy = metrics.accuracy_score(y_test,y_pred)
```
